```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
M0 = 1  # Initial mass (normalized)
lambda_ = 0.03  # Decay constant
t = np.linspace(0, 150, 300)  # Time values from 0 to 150 seconds
M = M0 * np.exp(-lambda_ * t)  # Exponential decay formula

# Plot the graph
plt.figure(figsize=(8,6))
plt.plot(t, M, 'b', linewidth=0.8)  # Blue curve with thin lines
plt.xlabel('Time, sec', fontsize=12)
plt.ylabel('Fractional Mass, M(t) / M(0)', fontsize=12)
plt.xlim(0, 150)  # Match x-axis limits
plt.ylim(0, 1)    # Match y-axis limits
plt.grid(False)  # No grid to match the reference image
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define time range
t = np.linspace(0, 500, 500)

# Define parameters for the function
A = 20      # Amplitude of the initial drop
alpha = 0.02  # Decay rate
omega = 0.06  # Frequency of oscillation
B = 25      # Long-term temperature limit
beta = 0.005  # Growth rate

# Compute temperature variation
T = A * np.exp(-alpha * t) * np.cos(omega * t) + B * (1 - np.exp(-beta * t))

# Plot the graph
plt.figure(figsize=(8,6))
plt.plot(t, T, 'b', linewidth=0.8)  # Blue curve
plt.xlabel('Time', fontsize=12)
plt.ylabel('Temperature', fontsize=12)
plt.title('Solutions at X = 1', fontsize=14)
plt.xlim(0, 500)
plt.ylim(0, 30)
plt.grid(True)

# Show plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import make_interp_spline

# Load the CSV file
file_path = "data.csv"  # Replace with your actual file path
df = pd.read_csv(file_path)
df = df.sort_values(by=df.columns[0])

# Round all values to 3 decimal places
df = df.round(3)
# Extract X and Y values
df = df.groupby(df.columns[0]).mean().reset_index()
x = df.iloc[:, 0].values  # Convert to NumPy array for interpolation
y = df.iloc[:, 1].values

# Generate a smooth range of X values
x_smooth = np.linspace(x.min(), x.max(), 300)  # 300 points for smooth curve

# Create spline interpolation
coeffs = np.polyfit(x, y, deg=10)  # 5th-degree polynomial fit
poly = np.poly1d(coeffs)

y_smooth = poly(x_smooth)

# Plot the smooth curve
plt.figure(figsize=(8, 5))
plt.plot(x_smooth, y_smooth, color='r', label="Polynomial Fit")
#plt.scatter(x, y, color='r', s=10, label="Original Data")  # Keep original points for reference
plt.xlim(0, 150)   # X-axis range from 0 to 1
plt.ylim(0, 1)
# Labels and title
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.title("Smooth Curve Using Cubic Spline Interpolation")
plt.legend()
plt.grid(False)

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load first CSV file
df1 = pd.read_csv("data.csv")
df1 = df1.sort_values(by=df1.columns[0])  # Sort by x values

# Load second CSV file
df2 = pd.read_csv("data2.csv")
df2 = df2.sort_values(by=df2.columns[0])  # Sort by x values

# Extract x and y values
x1, y1 = df1.iloc[:, 0], df1.iloc[:, 1]  # First column as x, second as y
x2, y2 = df2.iloc[:, 0], df2.iloc[:, 1]

# Create a figure with two subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 5))  # 1 row, 2 columns

# Plot first graph
axes[0].plot(x1, y1, linestyle='-', color='r', label="Data 1")
axes[0].set_title("Mass loss as a function of time for a polymer solution")
axes[0].set_xlabel("Time (s)")
axes[0].set_ylabel("Mass Ratio M(t)/M(0)")
axes[0].grid(False)

# Force axes to start at 0
axes[0].set_xlim(left=0)  # X-axis starts at 0
axes[0].set_xlim(right=150)  # X-axis starts at 0
axes[0].set_ylim(bottom=0)  # Y-axis starts at 0

# Plot second graph
axes[1].plot(x2, y2, linestyle='-', color='b', label="Data 2")
axes[1].set_title("Solutions at X = 1")
axes[1].set_xlabel("Temperature (C)")
axes[1].set_ylabel("Time (s)")
axes[1].grid(False)

# Force axes to start at 0
axes[1].set_xlim(left=0)  # X-axis starts at 0
axes[1].set_xlim(right=500)  # X-axis starts at 0
axes[1].set_ylim(bottom=0)  # Y-axis starts at 0

# Adjust layout and show the plots
plt.tight_layout()
plt.show()
```
