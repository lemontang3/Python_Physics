```python
import numpy as np
import matplotlib.pyplot as plt

# Define the time range (adjust as needed)
t = np.linspace(0, 10, 100)  # Time from 0 to 10 seconds

# Define the decay equation
k = 0.1054
M_t = np.exp(-k * t)

# Plot the data
plt.plot(t, M_t, label=r'$\frac{M(t)}{M(0)} = e^{-0.1054t}$')
plt.xlabel('Time (t) is (s)')
plt.ylabel('Fractional Mass, M(t) / M(0)')
plt.title('Exponential Decay of Fractional Mass Ratio')
plt.grid(True)
plt.legend()
plt.show()
```
