```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import pywt

# Load the image in grayscale
image_path = '/workspaces/Python_Physics/Images/PDVB Butanol 4% wt to vol (2).jpeg'  # Update the path to your image
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Apply Discrete Wavelet Transform (DWT)
coeffs = pywt.dwt2(image, 'haar')  # You can change 'haar' to other wavelet types
cA, (cH, cV, cD) = coeffs

# Visualize the original image and wavelet coefficients
plt.figure(figsize=(12, 8))

# Original Image
plt.subplot(2, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

# Approximation Coefficients
plt.subplot(2, 2, 2)
plt.imshow(cA, cmap='gray')
plt.title('Approximation Coefficients (cA)')
plt.axis('off')

# Horizontal Detail Coefficients
plt.subplot(2, 2, 3)
plt.imshow(cH, cmap='gray')
plt.title('Horizontal Detail Coefficients (cH)')
plt.axis('off')

# Vertical Detail Coefficients
plt.subplot(2, 2, 4)
plt.imshow(cV, cmap='gray')
plt.title('Vertical Detail Coefficients (cV)')
plt.axis('off')

# Display the plots
plt.tight_layout()
plt.show()
```
