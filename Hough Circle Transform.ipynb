Example on how to do Hough Transforms which would be necessary in further investigation on self-propelled droplet coalescence

```python
import matplotlib.pyplot as plt
import cv2 as cv
import numpy as np
import os
```

```python
def houghCirlceTransform():
    root = os.getcwd()
    imgPath = os.path.join(root,"/workspaces/Python_Physics/Images/PDVB Butanol 4% wt to vol (4).jpeg")
    img = cv.imread(imgPath)
    imgRGB = cv.cvtColor(img,cv.COLOR_BGR2RGB)
    imgGray = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
    imgGray = cv.medianBlur(imgGray,21)
    circles = cv.HoughCircles(imgGray,cv.HOUGH_GRADIENT,dp=1,minDist=100,param1=150,param2=15,minRadius=5,maxRadius=150)
    circles = np.uint16(np.around(circles))
    for circle in circles[0,:]:
        cv.circle(imgRGB,(circle[0],circle[1]),circle[2],(0,0,0),10)
    plt.figure()
    plt.imshow(imgRGB)
    plt.show()
if __name__ == "__main__":
    houghCirlceTransform()
```
